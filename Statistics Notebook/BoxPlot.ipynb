```python
import numpy as np
import matplotlib.pyplot as plt
```

A boxplot or sometimes called box-and-whisker plot
The boxplot conveys less information than a histogram, but it takes up less space and is so well
suited to compare several datasets

The histogram depends on the choice of intervals, we dont have to deal with that when using boxplot

```python
# BoxPlot Code
# Sample data: Fuel consumption (MPG) for 32 cars
mpg_values = [
    18, 15, 18, 16, 17, 15, 14, 14, 14, 15,
    15, 14, 15, 14, 22, 18, 21, 21, 10, 10,
    11, 9, 27, 28, 25, 25, 26, 27, 17, 16,
    15, 14
]

# Create the boxplot
plt.figure(figsize=(8, 5))
plt.boxplot(mpg_values, vert=True, patch_artist=True, 
            boxprops=dict(facecolor='skyblue', color='black'),
            medianprops=dict(color='red', linewidth=2))

# Add title and labels
plt.title('Boxplot of Fuel Consumption (MPG)', fontsize=14)
plt.xlabel('Miles per Gallon (MPG)', fontsize=12)

# Display the plot
plt.tight_layout()
plt.show()
```

the lower whisker shows the smallest number in the data
the upper whisker shows the largest number in the data
the median (red) half are larger and half are smaller
the first quartile (below red), a quarter of the data are smaller and 3 quarters are larger
the third quartile (above red), a point where 3 quarters are smaller and 1 quarter are larger

The scatterplot
    is used to depict data that come as pairs

```python
# Scatter PLOT Code example
# Generate random data for demonstration
np.random.seed(42)  # For reproducibility
years_of_education = np.random.randint(8, 23, 30)  # Random years of education between 8 and 22
base_salary = np.array(years_of_education) * 4 + 20  # Base salary grows linearly with education
noise = np.random.randint(-10, 10, 30)  # Add some variability to make it realistic
salary = base_salary + noise  # Final salary values

# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(years_of_education, salary, color='blue', edgecolor='black', alpha=0.7, s=100)

# Add labels and title
plt.title('Scatter Plot: Years of Education vs. Salary', fontsize=16)
plt.xlabel('Years of Education', fontsize=14)
plt.ylabel('Salary (in $1000s)', fontsize=14)

# Add grid
plt.grid(alpha=0.5)

# Show the plot
plt.tight_layout()
plt.show()
```

PROVIDING CONTEXT IS IMPORTANT
Statistical analyses tydpically compare the observed data to a reference.

One way to provide context is by using SMALL MULTIPLES. The compact design of the boxplot makes it well suited for this task.

Small multiples are a series of similar graphs or charts, arranged together, that share the same axes, scale, and design to allow for easy comparison across different subsets of data. They are often used to show trends, patterns, or relationships in multi-dimensional datasets.
