```python
For data that are qualitative (colors, car types, etc.)
    Use pie chart or a dot plot'


# Pie Chart for eyeballing fractional of the total a category responds to
# Dot plot for comparing frequencies of various categories
```

```python
# Code for Pie Chart
# Data for the pie chart
labels = ['Company A', 'Company B', 'Company C', 'Company D']
sizes = [25, 35, 20, 20]  # Percentages
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']  # Custom colors
explode = (0, 0.1, 0, 0)  # Highlight Company B

# Create the pie chart
plt.figure(figsize=(3, 3))  # Set the figure size
plt.pie(
    sizes, 
    labels=labels, 
    colors=colors, 
    explode=explode, 
    autopct='%1.1f%%', 
    startangle=140, 
    shadow=True
)

# Add a title
plt.title("Market Share Distribution")

# Display the pie chart
plt.show()
```

```python
# Code for Dot Plot
# Data
categories = ['Math', 'Science', 'English', 'History']
scores = [85, 90, 78, 88]

# Create the dot plot
plt.figure(figsize=(8, 5))
plt.scatter(scores, categories, color='blue', s=100, edgecolor='black')

# Add labels and title
plt.xlabel('Scores')
plt.ylabel('Subjects')
plt.title('Student Test Scores by Subject')

# Customize grid
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```

Bar Graph is essentially a dot plot put on its side.

Histogram is similar to bar graph
    it allows to use blocks with different widths 
    The area of the blocks are proportional to frequency.

```python
# Code for Bar Graph
# Data for the graph
students = ['Alice', 'Bob', 'Charlie', 'David', 'Ella']
assignments_completed = [10, 15, 8, 20, 12]

# Create the bar graph
plt.figure(figsize=(5, 2))
plt.bar(students, assignments_completed, color='skyblue', edgecolor='black')

# Add labels and title
plt.title('Assignments Completed by Each Student', fontsize=14)
plt.xlabel('Students', fontsize=12)
plt.ylabel('Number of Assignments Completed', fontsize=12)

# Add grid for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the graph
plt.tight_layout()
plt.show()
```

```python
# Code for Histogram
# Sample data: Ages of people
ages = [12, 23, 37, 45, 29, 18, 67, 35, 55, 43, 24, 21, 33, 40, 72, 25, 60, 38, 50, 28]

# Define the bins and their custom widths
bins = [0, 10, 20, 35, 50, 65, 80]  # Custom bins with varying widths

# Calculate the frequencies for each bin
frequencies, _ = np.histogram(ages, bins=bins)

# Define the bin centers and widths for plotting
bin_centers = [(bins[i] + bins[i + 1]) / 2 for i in range(len(bins) - 1)]
bin_widths = [bins[i + 1] - bins[i] for i in range(len(bins) - 1)]

# Plot the histogram manually using `plt.bar()`
plt.figure(figsize=(8, 5))
plt.bar(bin_centers, frequencies, width=bin_widths, color='skyblue', edgecolor='black', alpha=0.7)

# Add labels and title
plt.title('Age Distribution with Variable Bin Widths', fontsize=14)
plt.xlabel('Age Groups', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Add grid for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the graph
plt.tight_layout()
plt.show()
```

The percentage falling into a block can be figured without a vertical scale since the 
total area equals to 100%

But it is helpful to have a vertical scale (density scale). Its unit is "% per unit"

1. Density (crowding): The height of the bar tells how many subjects there are for one unit on the
horizontal scale.

2. Percentages (relative frequencies): Area = height x width

In the example density is not used.
It means there are 
two people ages 10-20
seven people ages 20-35
six people ages 35-50
three people ages 50-65

```python
# CODE for Histogram with density/crowding
# Sample data: Ages of people
ages = [12, 23, 37, 45, 29, 18, 67, 35, 55, 43, 24, 21, 33, 40, 72, 25, 60, 38, 50, 28]

# Define the bins and their custom widths
bins = [0, 10, 20, 35, 50, 65, 80]  # Custom bins with varying widths

# Calculate frequencies and densities
frequencies, _ = np.histogram(ages, bins=bins)
bin_widths = [bins[i + 1] - bins[i] for i in range(len(bins) - 1)]
#densities = frequencies / (bin_widths * len(ages))  # Density calculation // THIS ONE IS WRONG
densities = frequencies / (np.array(bin_widths) * len(ages))  # Density calculation

# Define the bin centers for plotting
bin_centers = [(bins[i] + bins[i + 1]) / 2 for i in range(len(bins) - 1)]

# Plot the density histogram
plt.figure(figsize=(5, 3))
plt.bar(bin_centers, densities, width=bin_widths, color='skyblue', edgecolor='black', alpha=0.7)

# Add labels and title
plt.title('Age Density Distribution', fontsize=14)
plt.xlabel('Age Groups', fontsize=12)
plt.ylabel('Density', fontsize=12)

# Add grid for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the graph
plt.tight_layout()
plt.show()
```
